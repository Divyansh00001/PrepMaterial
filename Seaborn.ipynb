# Joinplot

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data = df,kind = 'hex')
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data = df,kind = 'reg')
```

# **Pair Plot**


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='SepalLengthCm')
```

```python
sns.distplot(df['SepalLengthCm'])
```

```python
sns.distplot(df,kde=False,bins=10)
```

Count_Plot

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='day',data = df)
```

```python
sns.barplot(y='smoker',x='day',data=df)
```
