```python
#importing numpy and pandas
import pandas as pd
import numpy as np
```

```python
#dataframe - combination of columns and rows

df = pd.DataFrame(np.arange(0,20).reshape(5,4),index = ['Row1','Row2','Row3','Row4','Row5'],columns = ['column1','column2','column3','column4'])
df.head()
```

```python
#accessing the elements
#loc--> accessing element based on the name of row or column
df.loc['Row1']
```

```python
#iloc--> accessing element based on index (integer) of given row or column
df.iloc[1]
```

```python
# Check the type
# Dataseries- 1D data(either row or column)

type(df.loc['Row1'])
```

```python
df.iloc[:,:]
```

```python
df.iloc[1:,1:2]
```

```python
type(df.iloc[1:,0:])
```

```python
#convert Dataframes into array
df.iloc[:,1:].values
```
