# Matplotlib



```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
#illustrating through example
x = np.arange(10,30)
y = np.arange(30,50)
```

```python
#scatter plot
plt.scatter(x,y,c='b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Graph in 2D')
#save file
plt.savefig('testt.png')
```

```python
y = x*x
```

```python
#plot
plt.plot(x,y,'r-+')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--+')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'r-*')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
#using equations
x = np.arange(1,10)
y = 3 * x + 5 - x
plt.plot(x,y,c='r')
plt.title('Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0,4 * np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title('Sine-Wave')
```

```python

#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,np.pi*5,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)
plt.plot(x,y_sin,c='r')
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,c='g')
plt.title('Cosine')


```

```python
#Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y)
plt.bar(x2, y2, color = 'r')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
#Histogram

a = np.array([22,45,31,10,3,23,44,5,35,23,3])
plt.hist(a,color = 'r')
plt.title('histogram')
```

```python
#Boxplot
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist = False)
```

```python
#pieChart
labels ='python','c++','ruby','.Net'
sizes = [240,570,450,640]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0,0,0.1,0)
plt.pie(sizes,explode = explode,colors= colors,labels = labels,autopct ='%1.1f%%',shadow = False)
plt.title('PieChart')
plt.axis('equal')
```
